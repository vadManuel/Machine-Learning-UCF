## Poblem 1
Find three simple functions from R^2 to R and suitable points that are:
- a local minimum
- a local maximum
- neither a minimum nor maximum, but the gradient is zero at this point.

```python
class GradientDescent():
    def __init__(self, f, f_prime, lr, x_init, a_init, b_init):
        self.f = f
        self.f_prime = f_prime
        self.xs = [x_init]
        self.a = a_init
        self.b = b_init
        self.lr = lr
        # test

    def step(self, num_steps=1):
        for _ in range(num_steps):
            x = self.xs[-1]

            # one step of gradient descent
            x = x - self.lr * self.f_prime(x)

            # update
#             self.xs.append(x)
            tmp_xs = self.xs.copy()
            self.a = min(self.xs[-1], self.a)
            self.b = max(self.xs[-1], self.b)

    def plot(self):     
        fig, ax = plt.subplots(1, 1, figsize=(8, 8))
        ax.grid(True)

        # plot function
        x_axis = np.linspace(self.a, self.b)
        ax.plot(x_axis, [self.f(x) for x in x_axis])

        # plot all points encountered during gradient descent
        ax.scatter(self.xs, [self.f(x) for x in self.xs])
```

```python
def f(x):
    return 2.0 * (x - 2.0)**2

def f_prime(x):
    return 4.0 * (x - 2.0)

def g(x):
    return x ** 4 + 2 * (x - 0.1) ** 3 - x

def g_prime(x):
    return 4 * x ** 3 + 6 ** (x - 0.1) ** 2 - 1
```

```python
GD_f = GradientDescent(f, f_prime, lr=1.0, x_init=0.0, a_init=0.0, b_init=4.0)
GD_g = GradientDescent(g, g_prime, lr=0.05, x_init=1.0, a_init=-2.0, b_init=1.0)
```

```python
GD_f.step(2)
GD_f.plot()
```

```python
GD_g.step(2)
GD_g.plot()
```
